# Creamos una columna con el tipo de show

```python
# Creamos una lista con los open mics
open_mics = df_monologos_madrid[df_monologos_madrid['nombre'].str.contains('open mic', case=False)]['nombre'].tolist()
open_mics += df_monologos_madrid[df_monologos_madrid['nombre'].str.contains('Micro Abierto', case=False)]['nombre'].tolist()
open_mics += df_monologos_madrid[df_monologos_madrid['nombre'].str.contains('open', case=False)]['nombre'].tolist()


open_mics
```

```python
# Hacemos lo mismo con los concursos
concursos = df_monologos_madrid[df_monologos_madrid['nombre'].str.contains('concurso', case=False)]['nombre'].tolist()
concursos
```

```python
# Finalmente creamos la nueva columna con nuestras tres etiquetas, show, open mic y concurso
df_monologos_madrid['tipo_show'] = df_monologos_madrid['nombre'].apply(lambda x: 'open_mic' if x in open_mics else ('concurso' if x in concursos else 'show'))
df_monologos_madrid
```

```python
# Ahora sustituimos los valores y eliminamos la columna que nos sobra
df_monologos_madrid['lugar'] = df_monologos_madrid['lugar_modificada']
df_monologos_madrid = df_monologos_madrid.drop('lugar_modificada', axis=1)
df_monologos_madrid
```
