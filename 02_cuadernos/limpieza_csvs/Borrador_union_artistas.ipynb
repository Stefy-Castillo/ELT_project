```python
# Especificar la ruta y el nombre del archivo CSV
ruta_archivo_csv = 'C:/Users/STF/Desktop/Ironhack/proyectos/ETL_project/01_data/shows_provisional.csv'

# Cargar el archivo CSV en un DataFrame
df_monologos_madrid = pd.read_csv(ruta_archivo_csv)

df_monologos_madrid.head(10)
```

```python
df_unido = rec_artistas_1.merge(rec_artistas_2, on='nombre_show', how='inner')
df_unido
```

```python
condiciones = [
    (df_unido['artista_x'].isnull()) & (~df_unido['artista_y'].isnull()),
    (~df_unido['artista_x'].isnull()) & (df_unido['artista_y'].isnull()),
    (df_unido['artista_x'] == df_unido['artista_y']),
    (~df_unido['artista_x'].isnull()) & (~df_unido['artista_y'].isnull()) & (df_unido['artista_x'] != df_unido['artista_y'])
]

opciones = [
    df_unido['artista_y'],
    df_unido['artista_x'],
    df_unido['artista_x'],
    df_unido['artista_y']
]

df_unido['artistas_limpio'] = np.select(condiciones, opciones, default=np.nan)

df_unido
```
