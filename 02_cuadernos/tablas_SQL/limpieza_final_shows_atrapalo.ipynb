El Club de la Comedia es un show realizado por cómicos invitados de una forma bastante rotativa, por lo que asignamos el valor 'rotativo'.

Los otros dos shows tienen los nombres de los artistas en la descripción:

+ Summertime = Elena Beltrán, Miguel Campos, Yunez Chaib, Pablo Ibarburu, Helena Pozuelo, Iggy Rubín
+ Lgtb Comedy Jam = Albert Boira, Archie Bezos, Libertad Montero

```python
# Asignamos el valor correspondiente al club de la comedia

filtro = df_unido['nombre_show'] == 'La Hora y Media de El Club de la Comedia Madrid'
df_unido.loc[filtro, 'artista'] = 'rotativo'

# Definimos los valores que queremos asignar

summertime_artistas = ['Elena Beltrán', 'Miguel Campos', 'Yunez Chaib', 'Pablo Ibarburu', 'Helena Pozuelo', 'Iggy Rubín']
lgtb_comedy_jam_artistas = ['Albert Boira', 'Archie Bezos', 'Libertad Montero']

# Asignamos valores en función de la coincidencia de nombres de los listas
df_unido.loc[df_unido['nombre_show'].isin(['Summertime']), 'artista'] = ', '.join(summertime_artistas)
df_unido.loc[df_unido['nombre_show'].isin(['Lgtb Comedy Jam']), 'artista'] = ', '.join(lgtb_comedy_jam_artistas)

# Comprobamos
resultados3 = df_unido.loc[filtro2]
resultados3
```
