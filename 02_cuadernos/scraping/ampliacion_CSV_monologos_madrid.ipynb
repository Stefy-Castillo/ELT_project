```python
# Especificamos la ruta y el nombre del archivo CSV
ruta_archivo_csv = 'C:/Users/STF/Desktop/Ironhack/proyectos/ETL_project/01_data/artistas_madrid.csv'

# Cargamos el archivo 'artistas_madrid.csv' en un DataFrame
df_artistas = pd.read_csv(ruta_archivo_csv)

df_artistas.head(10)
```

```python
# Añadimos la nueva columna de artistas
df_merged = df_monologos_madrid.merge(df_artistas[['nombre', 'artista']], on='nombre', how='left')
df_merged.head(10)
```

```python
# Ahora sustituimos con la nueva columna 'artista_y' la anterior 'artista_x' a la cual le faltaban algunos artistas
df_merged.rename(columns={'artista_y': 'artista'}, inplace=True)
df_merged.drop('artista_x', axis=1, inplace=True)
df_merged.head()
```

```python
# Ponemos la columna donde le corresponde
columnas = ['nombre', 'precio', 'artista', 'lugar', 'nota', 'n_opiniones', 'duracion']
df_merged = df_merged[columnas]
df_merged.sample(20)
```

```python
# Especificamos la ruta y el nombre del archivo CSV
ruta_archivo_csv = 'C:/Users/STF/Desktop/Ironhack/proyectos/ETL_project/01_data/monologos_madrid_am.csv'

# Guardamos el nuevo DataFrame en un archivo CSV
df_merged.to_csv(ruta_archivo_csv, index=False)
```
