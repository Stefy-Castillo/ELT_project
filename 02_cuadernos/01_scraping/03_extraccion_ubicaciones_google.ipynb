```python
import pickle
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
# Especificamos la ruta y el nombre del archivo CSV
ruta_archivo_csv = 'monologos_madrid.csv'

# Cargamos el archivo CSV en un DataFrame
df_monologos_madrid = pd.read_csv(ruta_archivo_csv)
```

```python
# Comprobamos que hemos abierto correctamente el CSV
df_monologos_madrid.head()
```

```python
# Identificamos cuantos locales y qué locales tenemos

locales = df_monologos_madrid['lugar'].unique()
len(locales)
```

```python
# Hacemos una lista con los locales de los que queremos sacar la información

locales_lista = locales.tolist()
locales_lista
```

```python
# Librerias para realizar web scraping con selenium

from selenium import webdriver # webdriver permite manejar un navegador 
from webdriver_manager.chrome import ChromeDriverManager # permite instalar y mantener actualizado el driver de chrome
from selenium.webdriver.common.keys import Keys # permite simular teclas del teclado 
from selenium.webdriver.chrome.options import Options # permite configurar el driver de chrome como modo incognito o maximizar la ventana
from time import sleep # esperas entre ejecuciones de codigo


import warnings # para ignorar los warnings de python 
warnings.filterwarnings('ignore')
```

```python
# Entorno de prueba extracción

# Abre google
driver = webdriver.Chrome()
# Hacemos que vaya al buscador y espere a que cargue
driver.get("https://www.google.com/") 
driver.implicitly_wait(5)
# Hace click en el buscador
driver.find_element("css selector","#W0wltc > div").click()
sleep(2)

# Hacemos que busque el local y le de a ENTER
driver.find_element("xpath",'//*[@id="APjFqb"]').send_keys('Teatro Reina Victoria',Keys.ENTER)

try:
    puntuacion = driver.find_element("css selector", "span.Aq14fc").text
except:
    puntuacion = "No disponible"

try:
    N_reseñas = driver.find_element("css selector", "span.hqzQac a span").text
except :
    N_reseñas = "No disponible"

try:
    web_element = driver.find_element('css selector', 'a.ab_button[href^="http://"], a.ab_button[href^="https://"]')
    web = web_element.get_attribute('href')
except:
    web = "No disponible"

try:
    p_entradas_com = driver.find_element('css selector', '.inaKse.G5rmf').text
except:
    p_entradas_com = "No disponible"

try:
    o_entradas_com = driver.find_element('css selector', '.inaKse.KM6XSd').text
except:
    o_entradas_com = "No disponible"

try:
    p_facebook = driver.find_element('css selector', 'a:nth-child(2) > .inaKse.G5rmf').text
except:
    p_facebook = "No disponible"

try:
    o_facebook = driver.find_element('css selector', 'a:nth-child(2) > .inaKse.KM6XSd').text
except:
    o_facebook = "No disponible"

try:
    direccion = driver.find_element('css selector', 'span.LrzXr').text
except:
    direccion = "No disponible"
    
try:
    tipo_local = driver.find_element('css selector', 'div.zloOqf.kpS1Ac.vk_gy').text
except:
    tipo_local = "No disponible"

    
driver.quit()    
print(puntuacion)
print(N_reseñas)
print(web)
print(p_entradas_com)
print(o_entradas_com)
print(p_facebook)
print(o_entradas_com)
print(direccion)
print(tipo_local)
```

```python
# Extracción
locales_comedia_madrid = {"local":[],
                         "direccion": [],
                         "tipo_local": [], 
                        "valoracion": [],
                         "numero_reseñas": [],
                        "enlace_web": [], 
                        "puntuacion_entradas_com": [],
                   "numero_entradas_com": [],
                   "puntuacion_facebook": [],
                   "numero_puntuacion_facebook": []}

# Añadimos los nombres de los locales al diccionario:
locales_comedia_madrid["local"] = locales_lista

driver = webdriver.Chrome()
driver.get("https://www.google.com/") 
driver.implicitly_wait(5)

driver.find_element("css selector","#W0wltc > div").click()
sleep(2)

for local in tqdm(locales_lista):

    # Hacemos que busque el local y le de a ENTER
    driver.find_element("xpath",'//*[@id="APjFqb"]').send_keys(local,Keys.ENTER)

    
    try:
        locales_comedia_madrid["direccion"].append(driver.find_element('css selector', 'span.LrzXr').text)
    except:
        locales_comedia_madrid["direccion"].append(np.nan)

    try:
        locales_comedia_madrid["tipo_local"].append(driver.find_element('css selector', 'div.zloOqf.kpS1Ac.vk_gy').text)
    except:
        locales_comedia_madrid["tipo_local"].append(np.nan)
    
    try:
        locales_comedia_madrid["valoracion"].append(driver.find_element("css selector", "span.Aq14fc").text)
    except:
        locales_comedia_madrid["valoracion"].append(np.nan)

    try:
        locales_comedia_madrid["numero_reseñas"].append(driver.find_element("css selector", "span.hqzQac a span").text)
    except :
        locales_comedia_madrid["numero_reseñas"].append(np.nan)

    try:
        web_element = driver.find_element('css selector', 'a.ab_button[href^="http://"], a.ab_button[href^="https://"]')
        locales_comedia_madrid["enlace_web"].append(web_element.get_attribute('href'))
    except:
        locales_comedia_madrid["enlace_web"].append(np.nan)

    try:
        locales_comedia_madrid["puntuacion_entradas_com"].append(driver.find_element('css selector', '.inaKse.G5rmf').text)
    except:
        locales_comedia_madrid["puntuacion_entradas_com"].append(np.nan)

    try:
        locales_comedia_madrid["numero_entradas_com"].append(driver.find_element('css selector', '.inaKse.KM6XSd').text)
    except:
        locales_comedia_madrid["numero_entradas_com"].append(np.nan)

    try:
        locales_comedia_madrid["puntuacion_facebook"].append(driver.find_element('css selector', 'a:nth-child(2) > .inaKse.G5rmf').text)
    except:
        locales_comedia_madrid["puntuacion_facebook"].append(np.nan)

    try:
        locales_comedia_madrid["numero_puntuacion_facebook"].append(driver.find_element('css selector', 'a:nth-child(2) > .inaKse.KM6XSd').text)
    except:
        locales_comedia_madrid["numero_puntuacion_facebook"].append(np.nan)
        
        
    driver.find_element("css selector","#tsf > div:nth-child(1) > div.A8SBwf > div.RNNXgb > div > div.dRYYxd > div.BKRPef > div > span > svg").click()
    sleep(2)
    
driver.quit()  
```

```python
# Comprobamos las dimensiones

print(len(locales_comedia_madrid["local"]))
print(len(locales_comedia_madrid["valoracion"]))
print(len(locales_comedia_madrid["numero_reseñas"]))
print(len(locales_comedia_madrid["enlace_web"]))
print(len(locales_comedia_madrid["puntuacion_entradas_com"]))
print(len(locales_comedia_madrid["numero_entradas_com"]))
print(len(locales_comedia_madrid["puntuacion_facebook"]))
print(len(locales_comedia_madrid["numero_puntuacion_facebook"]))
```
